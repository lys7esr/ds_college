```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

sns.set_style("whitegrid") #graph style to whitegrid
```

```python
df=pd.read_csv("Data/Titanic-Dataset.csv")
df.head()
```

```python
df.shape
```

```python
df.tail()
```

```python
#histogram: numerical type (distribution of numerical data)
#kde: kernel distribution, smooth graph?
#age distribution
plt.figure(figsize=(8,5))
sns.histplot(df["Age"],bins=20,kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
#histogram for fare distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Fare'],bins=20,kde=True)
plt.title("Fare Distribution")
plt.show()
```

```python
#box plot (to find outliers) #outlier is the value out of the particular range the column is in
#dots in the graph depicts outlier
#left is min and right is max line

plt.figure(figsize=(6,4))
sns.boxplot(x=df['Age'])
plt.title("Age Boxplot")
plt.show()
```

```python
df.describe()
```

```python
#manually calculation
q1=20.125
q2=28
q3=38
iqr=q3-q1
lb=q1-(iqr*1.5)
ub=q3+(iqr*1.5)
print("Lower Bound: ",lb)
print("Upper Bound: ",ub)
```

```python
outlier_age=(df['Age']<lb) | (df['Age']>ub) #easy only
print(df[outlier_age])  #outlier_age is not a column so without quotes
```

```python
#bar plot: analyze single var in a range
#average fare by class
plt.figure(figsize=(7,5))
sns.barplot(data=df,x='Pclass',y='Fare',palette='viridis')
plt.title("Average Fare by Passenger Class")
plt.show()
```

```python
df.head()
```

```python
# survival rate by gender
plt.figure(figsize=(6,4))
sns.barplot(data=df,y='Survived',x='Sex',palette='Set2')
plt.title("Survival rate by gender")
plt.show()
```

```python
# count of no. of males and females
# categorical value alone cant plot a bargraph
# for bar graph need one categorical and one numerical
plt.figure(figsize=(6,4))
sns.countplot(data=df,x="Sex",palette='pastel') #for every category how much frequency and value
plt.title("No. of male and female")
plt.show()
```

```python
sex_count=df.groupby('Sex').size()
print(sex_count)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df,x="Pclass",palette='pastel') #for every category how much frequency and value
plt.title("No. of Pclass")
plt.show()
```

```python
#scatter plot
plt.figure(figsize=(6,5))
sns.scatterplot(data=df,x='Pclass',y='Fare')
plt.title("Pclass vs Fare")
plt.show()
```

```python
#line plot: find relation between two
plt.figure(figsize=(6,5))
sns.lineplot(data=df,x='Pclass',y='Fare')
plt.title("Pclass vs Fare lineplt")
plt.show()
```

```python
#pie chart #autopct is percentage format 1.1 %% second % is for ending 
column='Pclass'
plt.figure(figsize=(6,5))
df[column].value_counts().plot(kind='pie',autopct="%1.1f%%",startangle=90)
plt.title(f"{column} Distribution")
#plt.ylabel("")
plt.show()
```

```python
#correlation heatmap between numeric values
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='coolwarm') #annot=True to display the values in each block, cmap is for gradient
plt.title("Correlation Heatmap")
plt.show()
```

```python
# negative annotation means that the decrease i.e yk the normal graph decreases i.e. with pclass increasing the fare decreases
```

```python
#correlation matrix for missing values
plt.figure(figsize=(6,4))
sns.heatmap(df.isnull(),cbar=False,cmap='viridis') #cbar=false means tht sidebar above wont show, default it will show
plt.title("Missing Values")
plt.show()
```
