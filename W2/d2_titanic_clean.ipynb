```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Data/Titanic-Dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
#percentage of missing values
missing_percent=(df.isnull().sum()/len(df))*100
print(missing_percent)
```

```python
#fill missing values in numerical data with median
age_median=df['Age'].median()
df['Age']=df['Age'].fillna(age_median)
```

```python
df['Age'].isnull().sum() #numerical data use mean, median
```

```python
df['Age'].isnull().count
```

```python
#replace category data with mode
embarked_mode=df['Embarked'].mode()[0]
df['Embarked']=df['Embarked'].fillna(embarked_mode)
```

```python
#drop cabin column as most no. of null values
df=df.drop(columns=['Cabin'])
```

```python
df.isnull().sum() #to verify if there is any null value still present
```

```python
#DUPLICATED ROWS, to reduce redundancy, model trained on biased data
duplicate_rows=df.duplicated().sum()
print("Duplicated Rows: ",duplicate_rows) #no duplicate rows
```

```python
#if duplicates are present
#df=df.drop_duplicates()
```

```python
#Outlier box plot for age, fare
Q1=df['Fare'].quantile(0.25)
Q3=df['Fare'].quantile(0.75)
IQR=Q3-Q1
lb=Q1-1.5*IQR
ub=Q3+1.5*IQR
df=df[(df['Fare']>=lb)&(df['Fare']<=ub)] #to update df and remove all the outliers from fare column
```

```python
#check unique values
sex_unique=df['Sex'].unique()
print(sex_unique)
```

```python
#check unique values in embarked
embarked_unique=df['Embarked'].unique()
print(embarked_unique)
```

```python
#droping the columns that are not important, or yk just the use case
df=df.drop(columns=['PassengerId','Name','Ticket']) #if ran once then dont run again else error
```
